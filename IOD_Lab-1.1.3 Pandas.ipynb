# 4. Visualisation


#### 4.A Plot a histogram of `age`

````python
# A:
num_bins = 50
n, bins, patches = plt.hist(df.age, num_bins, facecolor='pink', alpha=0.8)
plt.show()
````

#### 4.B BoxPlot of `age`

````python
# A:
import matplotlib.pyplot as plt

plt.boxplot(df["age"])
plt.yscale("linear")
plt.grid(visible=True, which='major', color='g', linestyle='-')
plt.grid(visible=True, which='minor', color='b', linestyle='--')
plt.show()
````

#### 4.C Create a scatterplot of `age` and `length_of_service`

````python
# A:
plt.scatter(df.age, df.length_of_service, color="green", alpha=0.8)
plt.xlabel("Age")
plt.ylabel("Length of Service (Years)")
plt.title("Scatterplot of Age vs. Length of Service")
plt.show()
````

#### 4.D Print a statistical summary of `age`

````python
# A:
print("Statistical Summary of Age:", stats.describe(df.age))
````
